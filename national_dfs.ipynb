```python
import pandas as pd

def parse_tl_csv(filepath, skiprows=3):
    df = pd.read_csv(filepath, skiprows=skiprows)
    
    df = df.melt(
        id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
        var_name="Year",
        value_name="Value"
    )
    
    df["Year"] = pd.to_numeric(df["Year"], errors="coerce").astype("Int64")
    df = df.dropna(subset=["Value"])
    return df
```

```python
with open("national_data/consumer price idx/cpi.csv", "r") as f:
    for _ in range(10):
        print(f.readline())
```

```python
cpi_df = parse_tl_csv("national_data/consumer price idx/cpi.csv")
export_growth_df = parse_tl_csv("national_data/export growth/export_growth.csv")
gdp_growth_df = parse_tl_csv("national_data/gdp growth/gdp_growth.csv")
import_growth_df = parse_tl_csv("national_data/import growth/import_growth.csv")
unemployment_df = parse_tl_csv("national_data/unemployment/unemployment.csv")
```

```python
def rename_value(df, new_col_name):
    return df[["Country Name", "Year", "Value"]].rename(columns={"Value": new_col_name})
```

```python
cpi_df = rename_value(cpi_df, "Consumer price index (2010 = 100)")
export_df = rename_value(export_growth_df, "Exports of goods and services (annual % growth)")
gdp_df = rename_value(gdp_growth_df, "GDP growth (annual %)")
import_df = rename_value(import_growth_df, "Imports of goods and services (annual % growth)")
unemp_df = rename_value(unemployment_df, "Unemployment, total (% of total labor force) (annual % growth)")
```

```python
national_merged_df = cpi_df.merge(export_df, on=["Country Name", "Year"], how="inner")
national_merged_df = national_merged_df.merge(gdp_df, on=["Country Name", "Year"], how="inner")
national_merged_df = national_merged_df.merge(import_df, on=["Country Name", "Year"], how="inner")
national_merged_df = national_merged_df.merge(unemp_df, on=["Country Name", "Year"], how="inner")
```
